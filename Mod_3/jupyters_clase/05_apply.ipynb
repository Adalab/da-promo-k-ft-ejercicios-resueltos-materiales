# Groupby / apply

    - groupby -> lo usamos para agrupar por valores de una o varias columnas, lo usamos para entender nuestros datos y poder buscar patrones
    

```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../files/rating-and-performance.csv", index_col=0)
df.head()
```

```python
df.info()
```

```python
df.groupby("origin_country")["price"].mean().reset_index().sort_values(by="price", ascending=False)
```

```python
df.groupby("origin_country")["price"].mean().reset_index()
```

```python
df.groupby("origin_country")["price"].agg(["mean", "max", "min", "count"])
```

```python
df.groupby("origin_country").agg({"price":["min", "max"], "retail_price":["min","max"]}).reset_index()
```

```python
df.groupby("origin_country").agg({"price":["min", "max"]}).reset_index()
```

```python
df.groupby("origin_country")["price"].agg(["min", "max"]).reset_index()
```

```python
df.groupby("origin_country").mean(numeric_only=True)
```

```python
import numpy as np
```

```python
def cambio(dato):

    try:

        return float(dato.replace(",", ".").replace("$", ""))
    
    except:

        return np.nan
```

```python
cambio(valor)
```

```python
df["rate1"] = df["rate1"].apply(cambio)
df["rate2"] = df["rate2"].apply(cambio)
```

```python
lista_col = ["rate1", "rate2", ...]

for col in lista_col:

    df[col] = df[col].apply(cambio)
```
