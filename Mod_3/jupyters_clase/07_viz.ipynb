# Visualización

- Variables numéricas:
    - Histograma -> distribución de 1 variable
    - Boxplot -> distribución, incluye la mediana y podemos ver los valores atípicos (outliers)
    - Violinplot -> podemos ver la distribución (histograma) y la "caja" y "bigotes" (boxplot)

- Variables categóricas:
    - Countplot -> nos muestra la frecuencia, es el value_counts visto de forma más visual
    - Pieplot -> quesitos, tarta, ... proporción de una variable categórica

- Relaciones:

    - Dos variables numéricas:
        - scatterplot -> gráfica de puntos, nos muestra la posición de los valores de dos variables numéricas en un plano 2d. Podemos añadir hue para ver "una dimensión" más en nuestros datos.
        - regplot -> lo mismo pero con linea de regresión

    - Variable numérica y categórica:
        - Gráfico de barras -> barplot -> nos muestra la media para cada categoría en una barra
        - Boxplot -> vemos en detalle la comparación de cada grupo frente a una variable numérica
