- Métodos de unión: (uso práctico)

    - concat() -> lo vamos a usar para poner un df encima del otro (axis = 0)
        - "lo podemos entender como UNION en SQL"
        - las columnas se tienen que llamar igual para que aplile los datos
        - idnore_index = True para que genere un nuevo índice consecutivo y no use los valores originales

    - join / merge -> unen los df uno al lado del otro por el valor de columna/índice

        - join -> usamos los índices de los df. Índice 0 de df1 lo une con índice 0 de df2
            - por defecto hace un LEFT JOIN

        - merge -> igual que el join, pero usamos columnas columnas cómo unión -> fila id del df1 = "MCX" lo une con fila id del df2 = "MCX"
            - por defecto hace INNER JOIN

        - tipos de unión: "inner", "left", "right", "outer" -> dependiendo del tipo de unión nos puede generar valores nulos
