# EDA

- Exploración visual:
    - .head(n) -> primeros valores del df
    - .tail(n) -> úlyimos valores
    - .sample(n) -> valores aleatorios

- Exploración general:
    - .shape -> nú, de filas y columnas
        - [0] filas y [1] columnas
    - .columns -> nombre de las columnas -> podemos iterar
    - .select_dtypes() -> seleccionamos columnas por su tipo de dato
    - .describe() -> por defecto nos devuelve los estadísticos de las columnas numéricas. Si queremos incluir tipo object usamos include="O"
    - .info() ->  diferente información sobre el df
    - .unique() -> valores únicos de una columna
    - .value_counts() -> valores únicos y sus frecuencias, podemos verlo en % y decidir si queremos ver los valores nulos o no
