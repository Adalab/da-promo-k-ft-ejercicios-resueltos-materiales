### Ejercicios con loc e iloc

```python
# 1. Selecciona todas las columnas de las filas 2 a 4 (inclusive) utilizando loc
filas_2_a_4_loc = df.loc[2:4, :]
print("Filas 2 a 4 utilizando loc:")
filas_2_a_4_loc
```

```python
# 2. Utilizando loc, filtra el conjunto de datos para mostrar solo las filas donde la densidad (columna "Birth Rate") es mayor a 30
filtro_birth_rate_mayor_30 = df.loc[df['Birth Rate'] > 30, :]
print("\nFilas donde 'Birth Rate' es mayor a 30:")
filtro_birth_rate_mayor_30
```

```python
# 3. Utilizando loc selecciona las columnas "Country" y "Capital/Major City" de las filas donde el porcentaje de tierra agrícola (columna "Fertility Rate") es superior al 3
columnas_seleccionadas_fertility_rate = df.loc[df['Fertility Rate'] > 3, ['Country', 'Capital/Major City']]
print("\nFilas donde 'Fertility Rate' es mayor a 3 con columnas 'Country' y 'Capital/Major City':")
columnas_seleccionadas_fertility_rate
```

```python
# 4. Selecciona todas las filas de las columnas 0 a 4 (excluyendo la columna 4) utilizando iloc
columnas_0_a_4_iloc = df.iloc[:, 0:4]
print("\nTodas las filas de las columnas 0 a 4 utilizando iloc:")
columnas_0_a_4_iloc
```

```python
# 5. Encuentra el valor en la fila 2 y columna 7 utilizando iloc
valor_fila_2_columna_7_iloc = df.iloc[2, 7]
print("\nValor en la fila 2 y columna 7 utilizando iloc:")
valor_fila_2_columna_7_iloc
```

```python
# 6. Selecciona todas las filas para las columnas 0, 2, y 5 utilizando iloc
columnas_0_2_5_iloc = df.iloc[:, [0, 2, 5]]
print("\nTodas las filas para las columnas 0, 2, y 5 utilizando iloc:")
columnas_0_2_5_iloc
```

### BONUS DE LOC Y ILOC

```python
# 1. Encuentra el valor de "Gasoline Price" para el país cuya abreviatura es "AL" utilizando loc
precio_gasolina_al = df.loc[df['Abbreviation'] == 'AL', 'Gasoline Price'].values[0]
print(f"Precio de la gasolina para 'AL': {precio_gasolina_al}")
```

```python
# 2. Selecciona todas las columnas para el país cuya capital es "Tirana" utilizando loc
filas_tirana = df.loc[df['Capital/Major City'] == 'Tirana', :]
print("\nTodas las columnas para 'Tirana':")
filas_tirana
```

```python
# 3. Filtra el conjunto de datos para mostrar solo las filas donde el índice de CPI (columna "CPI") sea igual a 128.85
filas_cpi_128_85 = df.loc[df['CPI'] == "128.85", :]
print("\nFilas donde 'CPI' es igual a 128.85:")
filas_cpi_128_85
```

```python
# 4. Encuentra la densidad (columna "Density\n(P/Km2)") para el país cuya abreviatura es "AD" utilizando loc
densidad_ad = df.loc[df['Abbreviation'] == 'AD', 'Density\n(P/Km2)'].values[0]
print(f"\nDensidad para 'AD': {densidad_ad}")
```

```python
# 5. Selecciona las columnas "Country" y "Co2-Emissions"
columnas_pais_co2 = df[['Country', 'Co2-Emissions']]
print("\nColumnas 'Country' y 'Co2-Emissions':")
columnas_pais_co2
```

```python
# 6. Filtra el conjunto de datos para mostrar solo las filas donde el índice de CPI cambió un 2% (columna "CPI Change (%)")
filas_cpi_cambio_2 = df.loc[df['CPI Change (%)'] == "2", :]
print("\nFilas donde 'CPI Change (%)' es igual a 2%:")
filas_cpi_cambio_2
```

```python
# 7. Encuentra la tasa de natalidad (columna "Birth Rate") para el país con la abreviatura "AF" utilizando loc
tasa_natalidad_af = df.loc[df['Abbreviation'] == 'AF', 'Birth Rate'].values[0]
print(f"\nTasa de natalidad para 'AF': {tasa_natalidad_af}")
```

```python
# 8. Selecciona todas las columnas de las filas 3 a 5 (excluyendo la fila 5) utilizando iloc
filas_3_a_5_iloc = df.iloc[3:5, :]
print("\nTodas las columnas de las filas 3 a 5 (excluyendo la fila 5):")
filas_3_a_5_iloc
```

```python
# 9. Selecciona las columnas 1, 3 y 6 para todas las filas utilizando iloc
columnas_1_3_6_iloc = df.iloc[:, [1, 3, 6]]
print("\nColumnas 1, 3 y 6 para todas las filas:")
columnas_1_3_6_iloc
```
