```python
# Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde. 

def cambiar_tipo_dato(col):
    try: 
        col = col.replace(",", ".")
        return float(col)
    except:
        return np.nan
```

```python
lista_columnas = [ "land_area", "density", "cpi", "armed_forces_size", "co2-emissions", "population", "urban_population", "latitude", "longitude"]
for columna in lista_columnas:
    df[columna] = df[columna].apply(cambiar_tipo_dato)
```

```python
df["gdp"].unique()
```

```python
df["gdp"] = df["gdp"].apply(lambda x: float(x.replace(",", "") if type(x) == str else np.nan))
```

# Ejercicios groupby

```python
# 1. ¿Cuál es la densidad de población promedio por continente? 
df.groupby("continent")["density"].mean().reset_index()
```

```python
# 2. ¿Cuál es la esperanza de vida promedio por idioma oficial? Agrupa los países por idioma oficial y calcula la esperanza de vida promedio.
df.groupby("official_language")["life_expectancy"].mean().reset_index().sort_values(by = "life_expectancy", ascending = False).head()
```

```python
# 3. ¿Cuál es la tasa de mortalidad infantil promedio por moneda?
df.groupby("currency-code")["infant_mortality"].mean().reset_index()
```

```python
# 4. ¿Cuál es el salario mínimo por continente?
df.groupby("continent")["minimum_wage"].min().reset_index()
```

```python
# 5. ¿Cuál es la tasa de impuestos total máxima por densidad de población? Para solucionar este ejercicio puedes: 

df_impuestos = df.groupby("continent")[["total_tax_rate", "density"]].sum().reset_index()
df_impuestos["impuestos_total"] = df_impuestos["total_tax_rate"] / df_impuestos["density"]
df_impuestos
```

```python

# 8. ¿Cuál es la tasa de maternidad promedio por tasa de impuestos total? 

df_media_continente = df.groupby("continent")[['maternal_mortality_ratio', "total_tax_rate"]].mean()
df_media_continente["ratio"] = df_media_continente["maternal_mortality_ratio"] / df_media_continente["total_tax_rate"]
df_media_continente.sort_values(by = "ratio", ascending=False)
```

9.1. ¿Cuál es la tasa de natalidad máxima por área forestal? 

1. Primero, calculamos  la tasa de natalidad máxima y el area forestal máximo para cada continente. P

2. Luego, dividimos la tasa de natalidad máxima por el área forestal máxima para obtener la tasa de natalidad por punto porcentual de área forestal.


```python

df_nac_forest = df.groupby("continent")[['birth_rate', "forested_area"]].max()

df_nac_forest["ratio"] = df_nac_forest["birth_rate"] / df_nac_forest["forested_area"]
df_nac_forest.sort_values(by = "ratio")
```
