
4. Promedio de unidades vendidas por tipo de envío. Nuestro objetivo es analizar cómo se correlaciona el tipo de envío con las unidades vendidas. 5. Reemplazo de Valores usando `map()`. A lo largo del DataFrame tenemos algunas columnas cuyos valores son cero y uno, lo que podría llevar a confusión en su interpretación. 

```python
columnas = ["uses_ad_boosts", 'badge_local_product', 'badge_product_quality', 'badge_fast_shipping', 'shipping_is_express' ]

mapa = {0:"No", 1: "Si"}

for col in columnas:
    df[col] = df[col].map(mapa)
```

```python
df.head()
```

6. Calificación de la calidad del producto. Sobreescribe la columna `badge_product_quality`  utilizando `apply()` para asignar "Buena" a las filas donde 'badge_product_quality' sea "No" y "Excelente" donde sea "Si".

```python
df["badge_product_quality"] = df["badge_product_quality"].apply(lambda x: "Excelente"  if x == "Si" else "Buena")
df["badge_product_quality"].unique()
```

7. Cálculo de descuento promedio. Crea una nueva columna `discount_percentage` utilizando `apply()` que calcule el porcentaje de descuento promedio basado en las columnas 'price' y 'retail_price'. Después elimina las columnas de `average_discount` y `mean_discount` 

```python
def calcular_descuento(precio_v, precio_r):
    diferencia = precio_v - precio_r
    
    return round((diferencia * 100) / precio_r, 2)
```

```python
df["discount_percentage"] = df.apply(lambda dataframe: calcular_descuento(dataframe["price"], dataframe["retail_price"]), axis = 1)
df.drop(["average_discount", "mean_discount"], axis = 1, inplace = True)
```

```python
df.head()
```

9. Evaluación de la valoración del vendedor. Utiliza `apply()` para crear una nueva columna 'seller_reputation' donde los valores sean "Buena" si 'merchant_rating' es mayor o igual a 4, y "Regular" si es menor a 4. Después elimina la columna `merchant_rating`. 


```python
df["seller_reputation"] = df["merchant_rating"].apply(lambda x: "Buena" if x >= 4 else "Regular" )
df.drop("merchant_rating", axis = 1, inplace = True)
```
