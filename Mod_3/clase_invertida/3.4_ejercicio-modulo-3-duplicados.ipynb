```python
# importando las librerias!

import pandas as pd
```

```python
# 1- Abrid el fichero de `rating-and-performance` con el método que sea más correcto.
# Paso 1: Abrir el archivo y mostrar las 5 primeras filas del DataFrame
df = pd.read_csv('rating-and-performance.csv')
# Eliminar la primera columna
df = df.iloc[:, 1:]
df.head()
```

```python
# 2- Mostrad las 10 últimas filas.
df.tail(10)
```

```python
# 3- Mostrad 5 filas aleatorias del DataFrame.
df.sample(5)
```

```python
# 4- ¿Cuántas filas y columnas tenemos en el DataFrame?
print(f"tenemos {df.shape[0]} filas")
print(f"tenemos {df.shape[1]} columnas")
```

```python
# 5- ¿Cuáles son los nombres de las columnas del DataFrame?

print(f" las columnas de mi dataframe son {df.columns}")
```

```python
# 6- Seleccionad solo las columnas de tipo categórico del DataFrame. Después, mostrad los valores únicos y sus frecuencias para cada una de las columnas seleccionadas. 

categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
for column in categorical_columns:
    print("Valores únicos y frecuencias para", column)
    print(df[column].value_counts())
```

```python
#7- En el DataFrame tenemos dos columnas que parece que nos dan la misma información "title" y "title_orig". Eliminad "title". 

print(f"cantidad de columnas antes de eliminar la columna title: {df.shape[1]}")
df.drop(columns=['title'], inplace=True)
print(f"cantidad de columnas despues de eliminar la columna title: {df.shape[1]}")
print("DataFrame después de eliminar 'title':")
df.head()
```

```python
#8- ¿En que columnas tenemos valores nulos? Utilizad el método que más os guste. 

valores_nulos_por_columna = df.isnull().sum()

# Filtrar para mostrar solo las columnas que tienen valores nulos
columnas_con_nulos = valores_nulos_por_columna[valores_nulos_por_columna > 0]

columnas_con_nulos
```

```python
#9- ¿Tenemos valores duplicados en el DataFrame?

df.duplicated().sum()

#si, hay 34 duplicados!
```
