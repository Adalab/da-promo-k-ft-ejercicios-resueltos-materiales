```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df_insight = pd.read_csv("insight.csv")
df_insight.head(2)
```

```python
df_rating = pd.read_csv("rating-and-performance.csv")
df_rating.head(2)
```

```python
# 1. Unir los dos DataFrames en uno solo, usanado el método el método que considereis más correcto. 
df = df_rating.merge(df_insight, left_on = "merchant_id", right_on = "merchant.id", how = "inner", suffixes = ["_sold", "_employee"] )
df.head(2)
```

```python
df.columns
```

```python
df.drop(["index_sold", "index_employee", "title_orig", "merchant.id", "crawl_month", "theme", "currency_buyer"], axis = 1, inplace = True)
df.head(2)
```

```python
# 3- Algunas de las columnas tienen `.` en sus nombres y en otras tenemos `_`.  Unifica el nombre de las columnas para que todas tengan "_". 
nuevas_columnas = {col: col.replace(".", "_").lower() for col in df.columns}

df.rename(columns = nuevas_columnas, inplace = True)
df.head()
```

```python
#  Usando el método de limpieza, elimina de la columna "merchant_info_subtitle" la parte del *string* de los paréntesis. 

df["merchant_info_subtitle"]  = df["merchant_info_subtitle"].str.replace("\(.*\)", "", regex = True)
df.head()
```

```python
eliminar = ["badges_count", "product_variation_inventory",'has_urgency_banner',  "urgency_text", 'merchant_title', 'merchant_has_profile_picture',  'merchant_profile_picture', 'product_picture', 'total_urgency_count', 'urgency_text_rate']
df.drop(eliminar, axis = 1, inplace = True)
```

```python
df["merchant_info_subtitle"]  = df["merchant_info_subtitle"].replace("", np.nan, regex = True)
df.head(1)
```

```python
df.to_csv("merchant_fill.csv")
```

```python
# Seleccionad aquellos productos cuyo color es verde. ¿Cuántos productos tenemos?
df_verde  = df[df["product_color"] == "green"]
df_verde.shape
```

```python
# Seleccionad aquellos productos cuyo color es verde y rojo. ¿Cuántos productos tenemos de cada tipo de color?
colores = ["red", "green"]

df_colores = df[df["product_color"].isin(colores)]
df_colores["product_color"].value_counts()
```

```python
# Seleccionar aquellos productos de la talla M, que valgan más de 15. ¿Cuántos productos tenemos?
condicion1 = df["product_variation_size_id"] == "M"
condicion2 = df["price"] > 15

df_talla_precio = df[condicion1 & condicion2]
df_talla_precio.shape
```

```python
# Seleccionad aquellos productos cuyo precio este entre 1 y 2 (ambos inclusive).¿Cuántos productos tenemos?

df_baratos = df[df["price"].between(1,2, inclusive = "both")] 
df_baratos.shape
```
