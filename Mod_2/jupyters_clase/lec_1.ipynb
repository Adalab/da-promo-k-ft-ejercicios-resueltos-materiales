# APIs 

## Qué es una API:

    - Puente que necesitamos para comunicarnos y pedir información a BBDD externas
    - Es un conjunto de reglas que nos permite conectar aplicaciones

## Endpoint:

    - URL
    - "Cómo acceder a los distintos departamentos"

## Headers:

    - Etiquetas

## Parámetros:

    - Opciones que podemos pasar en nuestra petición

## Códigos error:

    - https://http.cat/

## Documentación de la API:

 - Información sobre la API (endpoints, parámetros, ...)
